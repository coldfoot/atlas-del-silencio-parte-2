```python
import geopandas as gpd
import pandas as pd
```

```python
# Keeps only Zona en Reclamación
zer = zer[zer.ID==24].reset_index()
```

```python
# Reads the other state level data
states = gpd.read_file("../output/finished-geojsons/level_1_results.geojson")
```

```python
# Checks if CRS matches
zer.crs == states.crs
```

```python
# Puts the columns in the zer file in the same pattern as the other states
zer = zer.rename(columns={
    "ID":"id",
    "ESTADO":"name"
}).drop(columns='index')
```
